## 박스 플롯(Box plot)

많은 데이터를 그림을 이용하여 집합의 범위와 중앙값을 빠르게 확인 가능하며 통계적으로 이상값이 있는지 빠르게 확인 가능합니다.

- 하위경계 : 제 1사분위-1.5*IQR
- 최솟값 : 하위경계 내의 관측치의 최솟값
- 제1사분위($Q_1$) : 자료들의 하위 25%의 위치
- 제2사분위($Q_2$, 중위수, 중앙값) : 자료들의 50%의 위치로 두꺼운 막대로 가시성을 높여서 표현
- 제3사분위($Q_3$) : 자료들의 하위 75% 위치
- 최댓값 : 상위 경계 내의 관측치의 최댓값
- 상위경계 : 제 3사분위+1.5*IQR
- 수염(Whiskers) : $Q_1$, $Q_3$로부터 IQR*1.5 내에 있는 가장 멀리 떨어진 데이터까지 이어진 선
- 이상값(Outlier) : 수염보다 바깥쪽에 존재한 데이터
- IQR : Q3-Q1, 50%의 데이터들이 흩어진 정도


```python
import matplotlib.pyplot as plt

plt.boxplot(과목['국어'])
plt.show()
```

```python
q3=과목.quantile(0.75)
q3
```

```python
q1=과목.quantile(0.25)
q1
```

```python
IQR=q3-q1
IQR
```

```python
def kor_outlier(df):
    kor_score = df['국어']
    if kor_score > q3['국어'] + 1.5 * IQR['국어'] or kor_score < q1['국어'] - 1.5 * IQR['국어']:
        return True
    else:
        return False

과목['국어_이상치여부'] = 과목.apply(kor_outlier, axis = 1) 
과목
```
