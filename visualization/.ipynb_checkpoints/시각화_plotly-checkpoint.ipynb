```python
import plotly.express as px
import numpy as np
import pandas as pd

x_ = np.arange(100)
y_ = x_ ** 2
fig = px.line(x=x_, y=y_)
fig.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go

# gapminder 데이터 셋 사용하기
korea_life = px.data.gapminder().query("country == 'Korea, Rep.'") # 한국인 기대 수명
korea_life
```

```python
fig = px.line(korea_life, x='year', y='lifeExp', title='기대수명')
fig.show()
```

```python
# 한국 GDP 데이터 셋
korea_gdp = px.data.gapminder().query("country == 'Korea, Rep.'")
korea_gdp
```

```python
fig = px.bar(korea_gdp, x='year', y='gdpPercap', title='한국인GDP')
fig.show()
```

```python
# 한국 GDP 데이터 셋
korea_data = px.data.gapminder().query("country == 'Korea, Rep.'")
korea_data
```

```python
fig = px.scatter(korea_data, x='gdpPercap', y='lifeExp')
fig.show()
```

```python
fig = px.pie(values=[50, 30, 20])
fig.show()
```

```python
labels = ['one', 'two', 'three']
values=[20, 30, 50]
fig = go.Figure(data = [go.Pie(values=values, labels=labels)])
fig.show()
```

```python
import plotly.graph_objects as go

labels = ['Oxygen','Hydrogen','Carbon_Dioxide','Nitrogen']
values = [4500,2500,1053,500]

fig = go.Figure(data=[go.Pie(labels=labels, values=values, pull=[0,0.2,0.5,0])])
fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure(go.Sunburst(labels=['Eve','Cain','Seth','Enos','Noam','Abel','Awan','Enoch','Azura'],
                            parents=['','Eve','Eve','Seth','Seth','Eve','Eve','Awan','Eve'],
                            values=[10,14,12,10,2,6,6,4,4],))

fig.update_layout(margin = dict(t=0, l=0, r=0, b=0))
fig.show()
```
