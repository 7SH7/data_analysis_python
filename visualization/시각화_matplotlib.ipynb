## subplot

- 여러개의 그래프를 하나의 그림에 나타내도록 함
- `plt.subplot(row, col, index)`의 형태임(콤마 없이 사용 가능)

```python
x = np.linspace(0, 10, 21)
y = x ** 2

# row * col matrix에서 index번째에 locate!
plt.subplot(131)
plt.plot(x, y, 'r')

plt.subplot(133)
plt.plot(x, y, 'g')

plt.show()
```

```python
x = np.linspace(0, 10, 21)
y = x ** 2

plt.subplot(131)
plt.plot(x, y, 'r')

plt.subplot(132)
plt.plot(x, y, 'g')

plt.subplot(133)
plt.plot(x, y, 'b')

plt.show()
```

```python
x = np.linspace(0, 10, 21)
y = x ** 2

plt.subplot(211)
plt.plot(x, y, 'r')

plt.subplot(212)
plt.plot(x, y, 'g')

plt.show()
```

```python
x = np.linspace(0, 10, 21)
y = x ** 2

plt.subplot(221)
plt.plot(x, y, 'r')

plt.subplot(222)
plt.plot(x, y, 'g')

plt.subplot(223)
plt.plot(x, y, 'b')

plt.subplot(224)
plt.plot(x, y, 'y')

plt.show()
```

```python
x = np.linspace(0, 10, 21)
y = x ** 2

x_ = np.linspace(0, 20, 41)
y_ = x_ ** 2

ax1 = plt.subplot(211)
plt.plot(x, y, 'r')
# plt.xticks(visible=False)

# ax2 = plt.subplot(212)   # x축 길이가 ax1과 공유되지 않음
ax2 = plt.subplot(212, sharex=ax1)
plt.plot(x_, y_, 'g')

plt.show()
```

```python
x = np.linspace(0, 10, 101)
y = x ** 2

plt.plot(x, y, 'r')
plt.grid(True)
# plt.grid(True, axis='x')
# plt.grid(True, axis='y')

plt.show()
```

```python
plt.subplot(231)
plt.plot(x, y, 'r')

plt.subplot(232)
plt.plot(x, y, 'g')

plt.subplot(233)
plt.plot(x, y, 'b')

plt.subplot(234)
plt.plot(x, y, 'r')

plt.subplot(235)
plt.plot(x, y, 'g')

plt.subplot(236)
plt.plot(x, y, 'b')

plt.show()
```

```python
# x * y matrix
# 이 코드는 2행 3열의 그리드를 생성합니다. 
# 각 서브플롯은 이 그리드의 특정 위치에 배치됩니다.
grid = plt.GridSpec(2, 3)  # 없어도 배치는 되는데, 명시성 부여여

plt.subplot(grid[0, 0])
plt.plot(x, y, 'r')

plt.subplot(grid[0, 1:])
plt.plot(x, y, 'g')

plt.subplot(grid[1, :2])
plt.plot(x, y, 'r')

plt.subplot(grid[1, 2])
plt.plot(x, y, 'g')

plt.show()
```

```python
# fig, axs : 각 subplot 사이 padding두겠다
# subplot 사이 간격 두려면 이 방식 필수!
fig, axs = plt.subplots(2, 3)
fig.subplots_adjust(wspace=.3, hspace=0.4)

plt.subplot(231)
plt.plot(x, y, 'r')
plt.title('title1')

plt.subplot(232)
plt.plot(x, y, 'g')
plt.title('title2')

plt.subplot(233)
plt.plot(x, y, 'b')
plt.title('title3')

plt.subplot(234)
plt.plot(x, y, 'r')
plt.title('title4')

plt.subplot(235)
plt.plot(x, y, 'g')
plt.title('title5')

plt.subplot(236)
plt.plot(x, y, 'b')
plt.title('title6')

plt.show()
```

```python
# 0부터 10까지 11개!
x = np.linspace(0, 10, 11)
y = x ** 2

plt.plot(x, y, 'k:', linewidth=3, marker='^', markersize=10, markerfacecolor='green')
plt.show()
```

```python
for m in ['+',',','.','1','2','3','4','o','x','+','v','^','<','>','s','d','p']:
  plt.plot(np.random.randint(20),
           np.random.randint(20),
           m,
           markersize=10,
           label = f'marker={m}'
           )

plt.legend(loc='upper center',bbox_to_anchor = (1.2,1.05),
           fancybox = True, shadow=True)

plt.show()
```

## bar chart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
label = ['one', 'two', 'three']
value = [10, 20, 30]
color = ['r', 'g', 'b']

plt.bar(label, value, color=color, width=0.5)
plt.title('hello', fontsize=20)
plt.xlabel('hellox', fontsize=15)
plt.ylabel('helloy', fontsize=15)
plt.show()
```

```python
label = ['one', 'two', 'three']
value = [-10, -20, 30]
color = ['r', 'g', 'b']

# bar height로!
plt.barh(label, value, color=color, height=0.5)
plt.title('hello', fontsize=20)
plt.xlabel('hellox', fontsize=15)
plt.ylabel('helloy', fontsize=15)
plt.show()
```

```python
plt.bar(['one', 'two', 'three'], [10, 20, 30], color='g')
plt.bar(['one', 'two', 'three'], [20, 40, 60], color='b', 
         alpha=0.5, bottom=[10, 20, 30])

plt.title('hello', fontsize=20)
plt.xlabel('hellox', fontsize=15)
plt.ylabel('helloy', fontsize=15)

plt.legend(['chart1', 'chart2'])
plt.show()
```

```python
plt.bar(['one', 'two', 'three'], [10, 20, 30], color='g')
plt.bar(['one', 'two', 'three'], [20, 40, 60], color='b', bottom=[10, 20, 30])

plt.title('hello', fontsize=20)
plt.xlabel('hellox', fontsize=15)
plt.ylabel('helloy', fontsize=15)

plt.legend(['chart1', 'chart2'])
plt.show()
```

### histogram

- 도수분포표를 그래프로 나타낸 방법이다.
- 막대그래프는 계급 즉 가로를 생각하지 않고 세로의 높이로만 나타낸다. (출처 : 위키백과)
- 연속형 자료를 계급으로 나누어  계급별 도수를 막대로 나타낸다.

```python
x = [np.random.randint(1, 7) for i in range(100000)]
# print(x)
```

```python
plt.hist(x, bins=11) #구간개수
# plt.xlim([0.5, 7.5])
plt.show()
```

```python
# 키를 조사합니다.
height = [173,182,163,175,192,189,190,200,176]
print(sorted(height))
plt.hist(height)
plt.show()
```

```python
# 키를 조사합니다.
height = [173,182,163,175,192,189,190,200,176]
plt.hist(height, bins=10, alpha=0.5)
plt.hist(height, bins=20, alpha=0.5)
plt.show()
```

```python
# 키를 조사합니다.
height = [173,182,163,175,192,189,190,200,176]
plt.hist(height, cumulative=True, label='one')
plt.hist(height, cumulative=False, label='two')

plt.legend(loc='upper left')
plt.show()
```

### pie chart

```python
labels = ['one', 'two', 'three']
# 100 / 100+20+10 만큼 차지 ...
size = [100, 20, 10]

plt.pie(size, labels=labels)
plt.show()
```

```python
labels = ['one', 'two', 'three']
size = [100, 20, 10]

plt.pie(size, labels=labels, shadow=True)
plt.show()
```

```python
labels = ['one', 'two', 'three']
size = [100, 20, 10]

plt.pie(size, labels=labels, shadow=True, autopct='%1.3f%%')
plt.show()
```

```python
labels = ['one', 'two', 'three']
size = [100, 20, 10]
explode = (0, 0.1, 0.3)  # 떨어진 정도

plt.pie(size, labels=labels, shadow=True, explode=explode, autopct='%1.3f%%')
plt.show()
```

```python
labels = ['one', 'two', 'three']
size = [50, 30, 20]
explode = (0, 0, 0)

# startangle = 시작점, counterclock : 방향 (false : 시계 방향
plt.pie(size, labels=labels, shadow=True, explode=explode, startangle=90, counterclock=False, autopct='%1.3f%%')
plt.show()
```
